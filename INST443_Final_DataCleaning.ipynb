# Data Cleaning and Manipulation

### OpenRefine 

The records we were able to retrieve using ABBYY FineReader were messy and inaccurate as seen in the image below. In order to clean the data and make it usable for data visualization and analysis, we utilized OpenRefine, an application for data cleanup and tranformation. 

Major Steps for Data Cleaning:
* Split column into multiple columns by seperator and rename them (Business Name or Owner, Address, and Category)
* Replace mistakes like "^" and "♦" with "*"
* Create Race column Based on Business Name or Owner column
* Replace unnecessary hyphens "-" in words with ""
* Replace mistakes like "15% e Boundary" with "15½ e Boundary"

Some General Refine Expression Language (GREL) Commands Used:
* value.replace("♦","\*")
* value.replace("^","\*")
* if(value.contains("*"), "Black", "")
* cells["Business Name or Owner"].value.contains(/[a-zA-Z]\*\-[a-zA-Z]\*/)
* replaceChars(cells["Business Name or Owner"].value, "-","")
* cells["Address"].value.contains(/[1-9]\*\%/)
* replaceChars(cells["Address"].value, "%","½")
 
Many more adjustments were needed than these in order to get to an end result of a cleaned and more accurate data set as found in the image below.
